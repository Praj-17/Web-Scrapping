# Web Scraping Using Beautiful soup

If you want to scrape a website 
1. Use the API
2. HTML web scraping using some tool llike bs4


```python
import requests
from bs4 import BeautifulSoup
url = "http://prajtech.xyz/"
```

# Now we are trying to get the html

```python
content= requests.get(url)

htmlContent = content.content
htmlContent
```

## Now the content is being loaded now we have to parse it

```python
soup = BeautifulSoup(htmlContent, 'html.parser')
soup.prettify
```

## Step 3: HTML Tree Traversal



```python
#Get the title of HTML PAGE
title = soup.title

print(type(soup))
print(type(title))
print(type(title.string))
title.string
```

Commonly Used types of objects
1. Tag - title
2. NavigableString - title.string
3. Beautifulsoup - soup
4. comment

```python
# Get all the paragraphs from the Page
paras = soup.find_all('p')
paras
```

```python
# Get all the anchor tags from the Page
anchors = soup.find_all('a')
for link in anchors:
    print(link.get('href'))
    # print(link.get('target'))
```

```python
#It will return the first that element
soup.find('p') 

```

```python
# Get classes of all the elements in the HTML PAGE
print(soup.find('p')['class'])

# For now it will throw error beacause there is no class in that P tag
```

```python
# find all the elements with the particular lead
print(soup.find_all('p', class_ = "lead"))
```


### Get the text from the tags/soup
