# Table Scrapper

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = "https://www.uob.com.sg/online-rates/gold-and-silver-prices.page"
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content, 'html5lib')
```

```python
table = soup.select('table#dataTablesFaculties')
```

```python
print(soup.prettify())
```

```python
tables = soup.find_all('tbody')
```

```python
len(tables)
```

```python
df = soup.find("tbody", {"id": "gld"})
```

```python
print(df)
```
