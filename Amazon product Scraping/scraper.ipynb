```python
import requests
import json
import pandas as pd
from autoscraper import AutoScraper
```

```python
data = pd.read_csv('Amazon Scraping.csv')
data.head()
```

```python
data = data.drop(index=0, axis = 1)
```

```python
urls = []
print(data['Asin'])
for asin, country in zip(data['Asin'], data['country']):
    url = f"https://www.amazon.{country}/dp/{asin}"
    urls.append(url)
urls
```
